- 과일을 먹는 순간 과일을 새롭게 생성된다

```python
import pygame
import time
import random

#초기화
pygame.init()

# 색상 설정
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
BLACK = (0,0,0)

# 화면 설정 (가로 800, 세로 600)
WIDTH, HEIGHT = 800, 600
win = pygame.display.set_mode((WIDTH, HEIGHT))

#글씨체 설정

#font = pygame.font.SysFont('arial', 20)
font = pygame.font.SysFont('kang.ttf', 20)
font_gameover = pygame.font.SysFont('kang.ttf', 72)

#게임 오버 폰트 및 텍스트 설정
game_over_text = font_gameover.render('Game Over', True, (0, 0, 255))  # 빨간색으로 설정
game_over_rect = game_over_text.get_rect(center=(WIDTH // 2, HEIGHT // 2))  # 화면 중앙에 위치


# 뱀 설정
snake_pos = [[100, 50], [90, 50], [80, 50]]
base_speed = 10
snake_speed = [base_speed, 0]

# 과일 설정
fruit_pos = [random.randrange(1, WIDTH//10)*10, random.randrange(1, HEIGHT//10)*10]
fruit_spawned = True

# 게임 설정
running = True
score = 0

while running:
    for event in pygame.event.get():
       if event.type == pygame.QUIT:
          running = False
       elif event.type == pygame.KEYDOWN:
          if event.key == pygame.K_UP:
             snake_speed = [0, -base_speed]
          elif event.key == pygame.K_DOWN:
             snake_speed = [0, base_speed]
          elif event.key == pygame.K_LEFT:
             snake_speed = [-base_speed, 0]
          elif event.key == pygame.K_RIGHT:
             snake_speed = [base_speed, 0]




    # 뱀의 움직임
    snake_pos.insert(0, list(snake_pos[0]))  # 먼저 뱀의 머리 부분을 복사             

    # 뱀이 움직이는 방향으로 머리 부분 이동
    snake_pos[0][0] += snake_speed[0]
    snake_pos[0][1] += snake_speed[1]

    # 뱀 성장 및 과일 생성
    if snake_pos[0] == fruit_pos:
        score += 1
        fruit_spawned = False
    else:
        snake_pos.pop()  # 뱀이 과일을 먹지 않았을 때만 꼬리 부분 제거 (머리가 복사되어 움직이기 때문에 꼬리 부분은 삭제해줌)

    if not fruit_spawned:
        fruit_pos = [random.randrange(1, WIDTH//10)*10, random.randrange(1, HEIGHT//10)*10]
        fruit_spawned = True

    # 벽 충돌 확인
    if snake_pos[0][0] >= WIDTH or snake_pos[0][0] < 0 or snake_pos[0][1] >= HEIGHT or snake_pos[0][1] < 0:
        running = False

    # 뱀 몸통 충돌 확인
    for pos in snake_pos[1:]:
        if snake_pos[0] == pos:
            running = False

   
    # 화면 그리기
    win.fill(WHITE) #화면 흰색으로 채우기


    #점수표현
    score_text = font.render(f'Score: {score}', True, (255, 0, 0))
    score_rect = score_text.get_rect(topright=(WIDTH - 20, 20))
    win.blit(score_text, score_rect)
    
    for pos in snake_pos:
        pygame.draw.rect(win, BLUE, pygame.Rect(pos[0], pos[1], 10, 10)) #win 창에 파란색 직사각형으로 x축 10, y축 10으로 직사각형을 그려줌 
    pygame.draw.rect(win, RED, pygame.Rect(fruit_pos[0], fruit_pos[1], 10, 10)) #win창에 빨간색 직사각형으로 x축 10, y축 10으로 직사각형을 그려줌

    print("현재 점수 : ",score)
    pygame.display.update()
    pygame.time.Clock().tick(30)
    


if not running:
    win.blit(, ) #텍스트, 위치
    pygame.display.update()  # 게임 오버 텍스트를 화면에 바로 업데이트
    time.sleep(3)  # 3초 동안 게임 오버 텍스트를 화면에 유지
    
pygame.quit()    
```
